```python
import requests
import bs4
```

```python
req1=requests.get('https://www.flipkart.com/apple-iphone-13-midnight-128-gb/p/itmca361aab1c5b0')
req1
```

```python
req1.content
```

```python
soup=bs4.BeautifulSoup(req1.text)
soup
```

```python
reviews=soup.findAll('div',{'class':'t-ZTKy'})
for rev in reviews:
    print(rev.get_text()+'\n')
```

```python
rating=soup.find('div',{'class':'_2d4LTz'})
print(rating.get_text())
```

```python
#all ratings
allrating=soup.findAll('div',{'class','_3LWZlK _1BLPMq'})
for allrate in allrating:
    print(allrate.get_text()+'\t\t')
```

```python
Customer_name=soup.findAll('p',{'class':'_2sc7ZR _2V5EHH'})
for all_name in Customer_name:
    print(all_name.get_text()+'\n')
```
